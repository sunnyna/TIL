# n_gram의 score 구하기

```
import numpy as np
import math 

training_set = np.array(df[['ngram+unigram', 'up_down']])

len(training_set)
```

```
training_set
```

```
from collections import defaultdict

doccnt1 = 0
doccnt0 = 0

wordfreq = defaultdict(lambda : [0,0])

pos = 0
neg = 0
neu = 0
# replace(';',',')
for doc, label in training_set:
  words = doc.replace('.',',').split(',')
  for word in words:
    if label == 'up':
      wordfreq[word][1] += 1
      pos += 1
    elif label == 'down':
      wordfreq[word][0] += 1
      neg += 1
    else:
      neu += 1

for key, (cnt0,cnt1) in wordfreq.items():
  if cnt1 > 0:
    doccnt0 += cnt0
  if cnt0 > 0:
    doccnt1 += cnt1

print('doccnt0 : {}'.format(doccnt0))
print('doccnt1 : {}'.format(doccnt1))    
```

```
wordfreq
```

```
hawkish_limit = 1.3
dovish_limit = 0.76

k = 0.5

polarity_score_list = []
polarity = []

wordprobs = defaultdict(lambda : [0,0])
for key, (cnt0, cnt1) in wordfreq.items():
  wordprobs[key][0] = (cnt0 + k) / (2 * k + doccnt0)
  wordprobs[key][1] = (cnt1 + k) / (2 * k + doccnt1)

  pos_score = wordprobs[key][1]
  neg_score = wordprobs[key][0]

  polarity_score = pos_score/neg_score
  polarity_score_list.append(polarity_score) 
  
  if polarity_score > hawkish_limit :
    polarity.append('Hawkish')
  if polarity_score < dovish_limit :
    polarity.append('Dovish') 


polarity_score_list
```
